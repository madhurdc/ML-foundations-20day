### Feature Engineering

#### BMI Category Normal/Overweight/Obese

```python
def bmiCategory(row):
    if row.bmi >= 30:
        return "Obese"
    elif row.bmi >= 25:
        return "Overweight"
    elif row.bmi >= 18.5:
        return "Normal"

df["BMI Category"] = df.apply(bmiCategory, axis="columns")
df
```

#### Age Group Senior/Adult/Young

```python
df["Age group"] = df.age.map(lambda row_age: "Senior" if row_age >= 60 else "Adult" if row_age >= 25 else "Young")
df
```

#### Create cost per child = charges / (children + 1)

```python
df["cost_per_child"] = df["charges"] / (df["children"] + 1)
df
```

#### Flag “high risk” = smoker AND BMI > 30

```python
df["High Risk"] = df.apply(lambda n: "Yes" if ((n["smoker"] == "yes") & (n["bmi"] > 30)) else "No", axis=1)
df
```

#### Which Demographic combination (region + sex + smoker) pays most?

```python
df.groupby(["region", "sex", "smoker"]).charges.max()
```
