```python
import pandas as pd
import numpy as np
```

```python
# DataFrames -> 2-Dimensional Data Structure (Rows + Columns)
```

```python
df = pd.DataFrame({
    "id" : [1,2,3],
    "age" : [23, 35, 19],
    "income" : [50000, 82000, 27000]
})
df
```

```python
# Series -> One-dimensional array with indexing. (NumPy Array with Index)
```

```python
# From a List
data = [10, 20, 30, 40]
mySeries = pd.Series(data)
mySeries
```

```python
# With a custom index
mySeries = pd.Series(data, index=['a', 'b', 'c', 'd'])
mySeries
```

```python
# From a Dictionary
data = {"apple": 10, "banana": 20, "cherry": 30}
series = pd.Series(data)
series
```

```python
# Each column of DataFrame is basically a Series
income = df["income"]
type(income)
```

```python
income.mean()
```

```python
# Exporting DataFrames to .csv or .xlsx files
```

```python
df = pd.DataFrame({
    "Name" : ["Alice", "Max", "John", "Alexa"],
    "Marks": [75, 86, 55, 100],
    "City": ["Nashik", "Mumbai", "Delhi", None]
})

df.to_csv("students.csv", columns=["Name", "Marks"], index=False)
```

```python
df.to_excel("students.xlsx", sheet_name="Marks", index=False)
```
