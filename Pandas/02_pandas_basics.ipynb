```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "ID" : [101, 102, 103, 104],
    "Name" : ["Alice", "Bob", "Max", "John"],
    "Age" : [25, 24, 30, None],
    "Salary" : [89000, 65000, 50000, 55000],
    "City" : ["Mumbai", "Pune", "Delhi", "Mumbai"]
})
df.set_index("ID", inplace=True)
```

```python
# COLUMN SELECTION
```

```python
# Selecting Columns
df["Age"].mean()
```

```python
# Selecting Multiple Columns
df[["Name", "Age"]]
```

```python
df_small = df[["Name", "Age", "City"]]
df_small
```

```python
# Selecting using dot notation
df.Age
```
